# 데이터 랭글링하기 위한 Pandas에서 지원하는 함수
- map
- apply
- applymap

```python
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.sex.map(lambda x: 1 if x == 'Male' else 2)
```

```python
tips.sex.apply(lambda x: 1 if x == 'Male' else 2)
```

axis 개념이 있어야 한다.

```python
tips.apply(lambda x: x['total_bill'] + x['tip'])
```

```python
tips.apply(lambda x: x['total_bill'] + x['tip'], axis=1)
```

```python
tips[['total_bill', 'tip']].apply(lambda x: x.sum())
```

```python
tips[['total_bill', 'tip']].apply(lambda x: sum(x))
```

```python
tips[['total_bill', 'tip']].applymap(lambda x: x + 1)
```

```python
tips['ratio'] = tips.apply(lambda x: x['tip'] / x['total_bill'], axis=1)
```
