```python
def recommendation_user(user_id, k, n=10):
    similar = reco_corr[user_id].sort_values(ascending=False)[1:1+k].index
    temp = pd.DataFrame(columns=["user_id", "item_id", "rating"])
    for i, j in enumerate(similar, start=1):
        no_watch = set(data[(data['user_id'] == j)].item_id) - set(data[data["user_id"] == user_id].item_id)
        k = data[(data['user_id']==j) & (data['item_id'].isin(no_watch)) & (data['rating']==5)]
        k.rating = k.rating * (1-(0.05*i)) # 유사도가 낮을수록 가중치 낮게 준다. # broadcasting
        temp = pd.concat([temp, k])
    temp = temp.sort_values('rating', ascending=False).drop_duplicates("item_id")[:n].rename(columns={"item_id":"movie_id"})
    result = temp.merge(item).drop_duplicates("movie_id")[["movie_title"]].reset_index(drop=True)
    return result
```
