```python
import requests
```

```python
# url 쪼개기
# urllib을 쓰지 않아도 requests 패키지에서도 지원함.
requests.compat.urlparse("http://pythonscraping.com/pages/cookies/login.html")
```

```python
# url 합치기
requests.compat.urljoin("http://pythonscraping.com/pages/cookies/login.html?id=1234", "process.php")
```

```python
import time

def download(method, url, param=None, data=None, headers = None, timeout=1, maxretries=3):
    try:
        resp = requests.request(method, url, 
                                params=param, data=data, 
                                headers = {'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/75.0.3770.100 Safari/537.36'})
        resp.raise_for_status()
    except requests.exceptions.HTTPError as e:
        if 500 <= e.response.status_code < 600 and maxretries > 0:
            print(maxretries)
            time.sleep(timeout)
            maxretries -= 1
            resp = download(method, url, param, data, headers, timeout, maxretries)
        else:
            print(e.response.status_code)
            print(e.response.reason)
    return resp
```

```python
# 아래 url은 로그인할때 쿠키가 생기는 것을 확인하기 위한 교육용 사이트
# 아무 아이디 패스워드를 넣어도 된다.

# 아이디와 비밀번호는 노출되면 안되기 때문에 post 방식을 사용함 (get방식은 url 주소에 파라미터가 노출됨.)
# 현재는 로그인 안됨.

url = "http://pythonscraping.com/pages/cookies/login.html"
url = requests.compat.urljoin(url, "welcome.php")
data = {"username":"1234", "password":"password"}

html = download("post", url, data=data)
html.text
```

```python
# cookie를 활용하기에 앞서서 session 객체를 만든다.

session = requests.Session()
```

```python
# url과 data를 날려서 response를 받아옴.
# 최초 접속해서 cookie 생성하고 로그인은 안된 상황

html = session.post(url, data)
html.text
```

```python
# 동일한 페이지에 다시 접속하면 쿠키값이 살아있어서 로그인 성공함.

html = session.post(url)
html.text
```

```python
# 쿠키 확인

session.cookies
```
