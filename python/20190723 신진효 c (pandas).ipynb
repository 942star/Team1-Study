```python
# 공공데이터 포털에서 아무데이터나 csv파일로 받아보자.
# local path에서 csv 읽어오기.
# 공공데이터 포털에서 받은 공공데이터는 잘 안열리는 경우가 많다.

# decodeError가 난다. => encoding 옵션을 주자.

pd.read_csv("대구광역시_서구_경로당현황_20171110.csv")
```

```python
# encoding='cp949' 를 주면 해결됨!
# 만약 OS 에러가 나면 engine='python'을 하면 거의 해결된다.

pd.read_csv("대구광역시_서구_경로당현황_20171110.csv", encoding='cp949')
```

```python
# 불러 들일때부터 index 컬럼을 정할 수 있다.

pd.read_csv("대구광역시_서구_경로당현황_20171110.csv", encoding='cp949', index_col="연번").head()
```

```python
# 불러들인 다음에 index를 바꿀 수 있다.

olds = pd.read_csv("대구광역시_서구_경로당현황_20171110.csv", encoding='cp949')

# index할 column을 지정해준다. 
# list로 여러개 column을 index로 지정할 수 있다.  ->  multi-index (나중에 배울거임)
olds.set_index("연번").head()
```

```python
# 바뀐게 적용되지 않았다.
# 반영시킬려면 재할당하는 방법과 inplace 방법이 있다.

olds.head()
```

```python
olds_raw = olds.copy()  # 원본은 따로 저장해둠. pandas의 copy는 numpy처럼 deepcopy이다.
olds.set_index('연번', inplace=True) # inplace는 mutable function으로 자기 자신을 바꾼다는 뜻이 된다.
olds.head()
```

```python
# reset_index는 새로운 인덱스를 새로 만든다.
# range index가 생긴다. (range한거처럼 숫자가 붙은 인덱스를 일컬음)

olds.reset_index()  .head()
```

```python
# drop=True는 기존에 있던 index를 지우고 인덱스를 새로 붙인다.

olds.reset_index(drop=True)  .head()
```

```python
# inplace까지 하면 자기 자신을 바꿈.

olds.reset_index(drop=True, inplace=True)
olds.head()
```

```python
olds = olds_raw.copy()
```

```python
# pandas에서 지우는건 모두 다 drop이라고 한다.

olds.drop(columns='연번', inplace=True)
olds.head()
```

```python
# drop에서 row를 지울 수도 잇다.
# 0번 컬럼 없어짐.

olds.drop(labels=0).head()
```

```python
# list로 묶으면 여러개 한번에 지울 수 있다.

olds.drop(labels=[1,3,5]).head()
```

```python
# axis의 기본값이 0 이어서 row가 삭제되었다.
# axis를 1로 바꾸면 column을 삭제할 수 있다.
# axis에 따라서 row 또는 column을 모두 지울 수 있어서 파라미터 이름이 labels 이다. (중의적)

olds.drop(labels='소재지', axis=1).head()
olds.head()
```

```python
# mapper 라는 parameter는 dictiorary를 받는다.

?olds.rename
```

```python
olds.rename({"소재지":"주소"}, axis=1, inplace=True)
olds.head()
```

```python
# 새로운 column 추가하는 방법
# 없는 column을 key로 주면 새로 생성한다.
# broadcating이 되므로 다음처럼 모양 안맞추고 쓸 수 있다.
# 우변에 컬럼을 주면 그 컬럼의 값을 복사한다.

olds['시'] = '대구광역시'
```
