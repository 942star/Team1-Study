```python
import pandas as pd
```

```python
# https://www.data.go.kr/dataset/15012005/fileData.do 에서 2018년 전국 상가업소 업력현황 다운로드

data = pd.read_excel("store.xlsx")
```

```python
data
```

```python
# 원본은 항상 보존해야한다.

data_raw = data.copy()
```

```python
# melt로 년도별 수를 합쳐주자.

data = data.melt(data.columns[:4], var_name="년", value_name="수")
```

```python
data_raw_na = data_raw.dropna()
```

```python
data = data_raw_na.melt(data.columns[:4], var_name="년", value_name="수")
```

```python
data.info()
```

```python
data
```

```python
# 수 column이 float dtpye으로 되어 있다.
# 수가 소수로 존재할리가 없다.
# int 타입으로 바꿔주자. (int라고 해도 되고 int64로 해야 될 수도 있다.)

data['수'] = data['수'].astype('int')
```

```python
data
```

```python
# 수가 0인 애들이 필요할까?
# 그건 테크닉에 따라서 다르다.
# 일단 0인 애들만 추려보자.

data['수'] == 0
```

```python
# 마스킹 기법
# != 는 '같지 않다' 라는 뜻
# 0이 아닌 애들만 마스킹한다.

data = data[data['수'] != 0]
data
```

```python
data.groupby('광역시도').count()
```

```python
# 업종 중 분류가 '연극/영화/극장' 인 것만 뽑기.

data[data.iloc[:,3]=="연극/영화/극장"]
```

```python
# 연극/영화/극장 의 광역시도 별 집계
# aggregation 분석 하는 방법이다.

# groupby 뒤에는 집계함수가 붙어야 한다.

data[data.iloc[:,3]=="연극/영화/극장"].groupby('광역시도').count()
```

```python
# 서울시의 부동산 업종만 뽑아보자.

data[data["광역시도"]=="서울특별시"][data.iloc[:,2]=="부동산"]
```

```python
# 서울시 구별로 부동산 집계를 뽑아보자.

data[data["광역시도"]=="서울특별시"][data.iloc[:,2]=="부동산"].groupby('시군구').count()
```

```python
# 조건을 &로 합칠 수 있다. 
# &로 조건을 합칠때는 항상 ()괄호로 묶어줘야 한다.
# 조건을 만든 다음에 fancy indexing

data[(data["광역시도"]=="서울특별시") & (data.iloc[:,2]=="부동산")].groupby('시군구').count()
```

```python
data[(data["광역시도"]=="서울특별시") & (data.iloc[:,2]=="부동산")].groupby(['시군구', '업종중분류']).count()
```

```python
# 시군구와 업종 중분류를 그룹으로 묶어서 가장 수가 많은 것 보기.

data[(data["광역시도"]=="서울특별시") & (data.iloc[:,2]=="부동산")].groupby(['시군구', '업종중분류']).max()
```

```python
# groupby가 뭔지 보자.

t = data[(data["광역시도"]=="서울특별시") & (data.iloc[:,2]=="부동산")].groupby(['시군구', '업종중분류'])
```

```python
# <>가 나왔다.
# 보통 이런 경우에는 irerator 였는데~~~ 얘도 iterator가 아닐까~~~?

t
```

```python
# next는 없다  =>  iterator는 아님.
# __iter__ 가 있다.  =>  순회가능한 녀석  =>  for 돌릴수 있다. 

dir(t)
```

```python
# groupby한 객체는 key와 value로 쌍을 이루고 있따.

for i,j in t:
    print(i,j)
```

```python
from collections import Counter
```

```python
# value의 종류별 갯수를 보여준다.

Counter('aaaaaabbbbbcbcccccc')
```

```python
# Series에서 value의 종류와 그 갯수를 뽑아준다.
# Counter와 비슷하다.
# python의 Counter를 이용해서 만들었다.

data["업종대분류"].value_counts()
```

```python
# pandas Series에도 Counter를 쓸 수 있다.
# 차이는 이쁘게 보이느냐, 딕셔너리로 만들었느냐.

Counter(data["업종대분류"])
```

```python
# 요렇게 묶어서 볼 수도 있다.
# 딕셔너리라서 가능함.

x = Counter(data["업종대분류"])
x.items()
```

```python
# most_common도 쓸 수 있다.

x.most_common()
```

```python
s = x.most_common()
```

```python
# 정렬하는 기능이다.
# key를 기준으로 한다. 
# 빈도 수를 기준으로 정렬해보자.

s.sort(key=lambda t:t[1], reverse=True)
```

```python
# 수가 큰 것부터 정렬되었다.

s
```

```python
# Counter처럼 dictionary를 주는게 아니기 때문에 sort나 내맘대로 가지고 놀기는 힘들다.
# 주어진 옵션 내에서는 내맘대로 할 수 있다. 
# normalize 옵션을 주면 전체 중의 비율로 계산해준다.

data["업종대분류"].value_counts(normalize=True)
```

```python
# groupby 3총사
# 1. pd.crosstab
# 2. pd.pivot_table
# 3. pd.groupby
```

```python
# 내가 원하는 데이터만 지정해서 볼 수 있다.
# 첫번째 인자가 index
# aggregation function을 선택할 수 있다.

pd.crosstab(data['광역시도'], data['년'], data['수'], aggfunc='count')
```

```python
# 순서를 바꾸면 table 모양이 바뀐다.
# 정렬을 년도로 할 수 있다.

pd.crosstab(data['년'], data['광역시도'], data['수'], aggfunc='count')
```

```python
pd.crosstab(data['년'], data['광역시도'], data['수'], aggfunc='count').plot.bar()
```

```python
# groupby로도 만들 수 있따.

data.groupby(['년', '광역시도']).count()
```

```python
data.groupby(['년', '광역시도']).count().unstack()
```

```python
data.pivot_table('수', '년', aggfunc='count')
```

```python
# 부분 합을 보여주는 margins 옵션

data.pivot_table('수', '년', aggfunc='count', margins=True)
```
